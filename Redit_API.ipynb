```python
import praw
import pandas as pd
```

```python
CLIENT_ID = "...."
CLIENT_SECRET = "...."
USER_AGENT = f"python:{CLIENT_ID}:0.1 (by /u/...)"
```

```python
api = praw.Reddit(
    client_id = CLIENT_ID,
    client_secret = CLIENT_SECRET,
    user_agent = USER_AGENT
)
```

```python
api.read_only
```

```python
items = []
submissions = api.subreddit("Utrecht").hot(limit = 200)

for submission in submissions:
    item = {
        "title" : submission.title,
        "score" : submission.score,
        "comments" : submission.num_comments
        
    }
    items.append(item)
```

```python
df = pd.DataFrame(items)
df
```

```python
df.head()
df.plot(kind='bar')
```
